# Analysis of the Titanic Dataset

## Introduction

The sinking of the RMS Titanic is one of the most infamous shipwrecks in history. On April 15, 1912, during her maiden voyage, the Titanic sank after colliding with an iceberg, resulting in the deaths of 1,502 out of 2,224 passengers and crew. This Jupyter Notebook aims to analyze the Titanic dataset to uncover insights about the passengers and the factors that influenced their survival.

The dataset contains various features such as passenger class, age, gender, fare, and survival status. Through visualizations and statistical analysis, we will explore the following questions:

1. What is the survival rate by passenger class?
2. How does the survival rate vary by gender within each passenger class?

By answering these questions, we aim to understand the impact of socioeconomic status and gender on the survival rates of the Titanic passengers.


**Summary of the Titanic Dataset:**

The Titanic dataset contains information about the passengers aboard the RMS Titanic, which sank on its maiden voyage in April 1912. The dataset includes variables such as:

- Passenger Class (Pclass): A categorical variable representing the class in which the passenger traveled (1st, 2nd, or 3rd class).
- Survived: A binary variable indicating whether the passenger survived (1) or did not survive (0).
- Sex: The gender of the passenger (male or female).
- Age: The age of the passenger in years.
- Fare: The fare paid for the ticket.
- Embarked: The port of embarkation (C = Cherbourg; Q = Queenstown; S = Southampton).

The data is valuable for understanding the passengers' demographics and how different factors, such as class, gender, and fare, influenced survival rates.

**Questions to Explore Visually:**

1. What is the survival rate by passenger class?
   Visualized in the Bar Graph for Survival Rate by Passenger Class: This graph shows how survival rates differ across the three passenger classes (1st, 2nd, and 3rd). It helps to identify if being in a higher class (e.g., 1st class) increased the chances of survival compared to being in a lower class (e.g., 3rd class).

2. How does the survival rate vary by gender within each passenger class?
   Visualized in the Facet Grid for Survival Rate by Class and Gender: This set of plots shows survival rates by class, split by gender. It allows us to explore the impact of both gender and class on survival rates. For instance, we can see if women in the first class had a higher survival rate than men in the same class or how survival rates differed for males and females in the third class.

These visual explorations provide insights into how socioeconomic status (as indicated by passenger class) and gender influenced survival outcomes on the Titanic. By answering these questions, we can better understand the factors contributing to a passenger's likelihood of surviving the disaster.

```python
# Histogram for Age Distribution

# Create a new figure for the plot with a specified size (10 inches wide by 6 inches tall).
plt.figure(figsize=(10, 6))

# Generate a histogram for the 'Age' column from the dataset 'df'. 
# The dropna() function is used to remove any missing values (NaN) from the 'Age' column 
# 'bins=20' specifies the number of bins to divide the age range into.
# 'kde=True' adds a Kernel Density Estimate (KDE) plot over the histogram, 
# which provides a smooth estimate of the distribution's density.
sns.histplot(df['Age'].dropna(), bins=20, kde=True)

# Set the title of the histogram to describe what is being plotted.
plt.title('Distribution of Passenger Ages')

# the x-axis  indicate the 'Age' of passengers.
plt.xlabel('Age')

# the y-axis  indicate that it represents the 'Frequency' 
plt.ylabel('Frequency')

# Display the plot on the screen.
plt.show()
```

```python
# defines the size of the plot in inches (width x height).
plt.figure(figsize=(10, 6))

# Create a scatter plot to visualize the relationship between Age and Fare
# x='Age' assigns passenger age to the x-axis.
# y='Fare' assigns the fare paid by passengers to the y-axis.
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')

# Add a title to the plot
plt.title('Bivariate Plot of Age and Fare')
# The title gives context to the plot, indicating that it shows the relationship between passenger age and fare.

# Label the x-axis
# This label indicates that the horizontal axis represents the age of the passengers.
plt.xlabel('Age')

# Label the y-axis
# This label indicates that the vertical axis represents the fare paid by passengers.
plt.ylabel('Fare')

# Display the plot
# plt.show() renders the plot in the output, making it visible for analysis.
plt.show()
```

**1. Survival Rate by Passenger Class:**

- The bar graph showing survival rates by passenger class reveals that the first class had the highest survival rates, followed by those in the second class, with the third class having the lowest survival rates. **2. Survival Rate by Gender within Each Class:**

- The facet grid plots, which split the data by both gender and passenger class, show that women generally had higher survival rates across all classes compared to men. This pattern is particularly evident in the 1st and 2nd classes, where most female passengers survived. In contrast, male passengers in the 3rd class had the lowest survival rates, highlighting the combined impact of both gender and class on survival chances.

**Conclusion:**
The analysis of the Titanic dataset concludes that passenger class and gender were significant factors in determining survival during the Titanic disaster. Specifically:

- **Higher Passenger Class:** Passengers in the 1st class had the highest likelihood of survival, likely due to their proximity to lifeboats and preferential treatment during the evacuation process.
- **Gender:** Female passengers, particularly those in the 1st and 2nd classes, had a much higher chance of survival than their male counterparts, likely due to the "women and children first" protocol followed during the evacuation.

**How Visualizations Support the Conclusion:**

- **Bar Graph for Survival Rate by Passenger Class:** This graph visually demonstrated the significant difference in survival rates across the different classes, clearly showing that 1st class passengers had the highest survival rate.
- **Facet Grid for Survival Rate by Gender and Class:** These plots allowed for a detailed examination of how survival rates varied by class and gender within each class. The visual distinction between the survival rates of males and females within each class supported the conclusion that gender and class were crucial in determining survival.

Using these visualizations, we can conclude that the social structure aboard the Titanic, as reflected in the class system and gender-based evacuation practices, played a pivotal role in determining the passengers' fates. The visual evidence provides a clear and compelling narrative of how these factors influenced survival outcomes.
